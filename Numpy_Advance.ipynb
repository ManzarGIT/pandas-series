# Boolean Indexing

```python
#  method used to select elements from a NumPy array based
# on a boolean mask (a boolean array of the same shape as the original array).
```

```python
b=np.random.randint(1,100,24).reshape(6,4)
b
```

```python
#find all the number greate than 50
b[b>50]  #using array [b>50] as mask
```

```python
#find all the number grater than and even
b[ b%2==0]
```

```python
b[(b>50) & (b%2==0)]   #  bitwise & used in boolean operator
```

```python
#not divisible by seven
b[~(b%7==0)]
```

# Broadcasting
The term broadcasting describe how numpy treats array with different shape during arithmetic operation .the smaller array is "broadcast" across the larger array so that they have compatible shapes.

# Broadcasting Rule
To broadcast two arrays together, NumPy compares their shapes element-wise from right to left, and two dimensions are compatible when:
*  They are equal, or
*  One of them is 1
If these conditions are met for all dimensions, NumPy broadcasts the smaller array to match the shape of the larger one.

If not, it raises a **ValueError**.

# Working with mathematical formula

```python

#sigmoid
def sigmoid(array):
  return 1/(1+np.exp(-array))

arr=np.arange(10)
sigmoid(arr)
```

```python
#mean squared error
actual=np.random.randint(1,50,25)
predicted=np.random.randint(1,50,25)
print(actual)
print(predicted)

def mse(actual,predicted):
  return np.square(np.subtract(actual,predicted)).mean()

print(mse(actual,predicted))

#categorical cross entropy
```

# Working with missing value


```python
a=np.array([1,2,3,4,5,np.nan,6,7])

np.isnan(a)     #to check is there any missing value

a[~np.isnan(a)] # to remove the missing value usnig boolean arrray operator
```
